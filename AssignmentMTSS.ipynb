(1) Prove that (AB)C = A(BC)

Let A,B and C be the matrix $a_{ij}, b_{ij}$ and $c_{ij}$. Then  multiplication of two matrices is given $$(AB)_{ij} = \sum_{k}(A_{ik}B_{kj})$$

$$((AB)C)_{rs} =  \sum_{w}(AB_{rw}C_{ws})$$

$$((AB)C)_{rs} =  \sum_{w}(\sum_{k}(A_{rk}B_{kw})C_{ws})$$

The left hand side can be written as, 

$$(A(BC))_{rs} =  \sum_{w}(A_{rw}(BC)_{ws})$$

$$(A(BC))_{rs} =  \sum_{w}\sum_{k}(A_{rw}(B_{wk}C_{ks}))$$

$$(A(BC))_{rs} =  \sum_{w}\sum_{k}(A_{rw}B_{wk})C_{ks})$$

$$(A(BC))_{rs} =  \sum_{w}\sum_{k}(AB)_{rk}C_{ks}) = (AB)C$$

$$ Hence Proved $$
