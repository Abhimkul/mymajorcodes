```python
import numpy as np
import scipy 
import matplotlib.pyplot as plot
import sklearn
from sklearn.cluster import KMeans
```

```python
#Seed
np.random.seed(1)
#data
data_x = np.random.random(200)
data_y = np.random.random(200)

#Plotting
plot.scatter(data_x,data_y)
plot.xlabel("x")
plot.ylabel("y")
```

```python
#K-Means Clustering 
k_mean = KMeans(n_clusters=3, random_state = 10, n_init=10, init= "k-means++")
data_1 = np.column_stack((data_x,data_y))

#Algorithm
k_clus = k_mean.fit_predict(data_1)
centroid = k_mean.cluster_centers_
data_new = np.column_stack((data_1,k_clus))
data_new

A = np.array([i for i in data_new if i[-1] == 0])
B = np.array([i for i in data_new if i[-1] == 1])
C = np.array([i for i in data_new if i[-1] == 2])

#Plotting
plot.scatter(A[:,0], A[:,1], color = "red")
plot.scatter(B[:,0], B[:,1], color = "blue")
plot.scatter(C[:,0], C[:,1], color = "green")
plot.scatter(centroid[:,0],centroid[:,1], color = "black",marker = "+")
plot.title("k=3")
```

```python
# For K = 4
np.random.seed(2)

#Data
data_x_1 = np.random.random(500)
data_y_1 = np.random.random(500)

#Plotting
plot.scatter(data_x_1,data_y_1)
plot.xlabel("x")
plot.ylabel("y")
```
