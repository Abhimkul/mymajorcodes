```python
import numpy as np
import matplotlib.pyplot as plot
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Seeding
np.random.seed(1)

#Data
data_1 = np.random.random(200)
data_2 = np.random.random(200)

data_ = np.column_stack((data_1,data_2))

#Target
target = np.random.choice([0,1], size = 200)

#Classification
plot.scatter(data_1,data_2)
plot.xlabel("x")
plot.ylabel("y")
```

```python
# K-Nearest neighbour classification
k_nn = KNeighborsClassifier(n_neighbors = 3)
k_nn.fit(data_,target)
```

```python
#Classification
point = np.array([[0.5,0.3]])
print(k_nn.predict(point))
```
